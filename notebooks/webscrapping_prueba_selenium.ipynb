```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import os

# Configuración de Selenium
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ejecutar sin abrir el navegador (opcional)
chrome_options.add_argument("--disable-gpu")

# Ruta al chromedriver
driver = webdriver.Chrome()
driver = webdriver.Chrome(service=Service(driver), options=chrome_options)

# URL inicial de la búsqueda
url = "https://www.fotocasa.es/es/comprar/viviendas/manises/todas-las-zonas/l"

# Carpeta para guardar los archivos HTML
output_folder = "../data/raw/fotocasaPrueba"
os.makedirs(output_folder, exist_ok=True)

# Abrir la página inicial
driver.get(url)
time.sleep(5)  # Esperar a que la página cargue completamente

# Contador para los archivos
page_counter = 1

while True:
    # Guardar el HTML de la página actual
    html_content = driver.page_source
    file_path = os.path.join(output_folder, f"fotocasaManisesPrueba{page_counter}.html")
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(html_content)
    print(f"Guardado: {file_path}")
    
    # Buscar el botón de "Siguiente"
    try:
        next_button = driver.find_element(By.XPATH, '//a[@aria-label="Siguiente"]')
        # Comprobar si el botón está habilitado
        if not next_button.is_enabled():
            print("No hay más páginas. Finalizando...")
            break
        # Hacer clic en el botón "Siguiente"
        next_button.click()
        time.sleep(5)  # Esperar a que la siguiente página cargue
        page_counter += 1
    except Exception as e:
        print(f"Error al navegar a la siguiente página: {e}")
        break

# Cerrar el navegador
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os

# Configuración de Selenium
chrome_options = Options()
# chrome_options.add_argument("--headless")  # Ejecutar sin abrir el navegador (opcional)
chrome_options.add_argument("--disable-gpu")

# Crear una instancia del navegador
driver = webdriver.Chrome(options=chrome_options)

# URL inicial de la búsqueda
url = "https://www.fotocasa.es/es/comprar/viviendas/manises/todas-las-zonas/l"

# Carpeta para guardar los archivos HTML
output_folder = "../data/raw/fotocasaPrueba"
os.makedirs(output_folder, exist_ok=True)

# Abrir la página inicial
driver.get(url)
time.sleep(5)  # Esperar a que la página cargue completamente

# Contador para los archivos
page_counter = 1

while True:
    # Guardar el HTML de la página actual
    html_content = driver.page_source
    file_path = os.path.join(output_folder, f"fotocasaManises{page_counter}.html")
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(html_content)
    print(f"Guardado: {file_path}")
    
    # Buscar el botón de "Siguiente"
    try:
        # Hacer scroll hasta el final de la página
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Esperar tras el scroll
        
        # Esperar hasta que el botón "Siguiente" sea visible y clicable
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//a[@aria-label="Siguiente"]'))
        )
        # Hacer clic en el botón "Siguiente"
        next_button.click()
        time.sleep(5)  # Esperar a que la siguiente página cargue
        page_counter += 1
    except Exception as e:
        print(f"No hay más páginas o error al navegar: {e}")
        break

# Cerrar el navegador
driver.quit()
```
